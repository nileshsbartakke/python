# Before Cleaning data Data Is coming from A Real Estate Domain So feature selection Is best choice to do analysis 

```python
#Subseting and creatig feature Matrix
feature_matrix=dataset.drop(['First Name', 'Last Name','Email 4','Email 4 - Type','Email 5','Email 5 - Type','Email 6','Email 6 - Type'], axis=1)
feature_matrix
```

# If Colume contain all na value then Drop colume value  and create another feature Matrix

```python
feature_matrix = dataset.dropna(axis = 1, how = 'all')
feature_matrix.count()
```

# Drop na value if colume contain 80% of na value

```python
new_feature1=feature_matrix[feature_matrix.columns[feature_matrix.isnull().mean() < 0.8]]
```
